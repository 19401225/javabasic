```python
# AI(Artificial Intelligence, 인공지능)

# 넓은 의미에서 인간의 합리적인 사고나 행동을 모방해 자동화한 프로세스
# 간단하게 생각하면 숫자와 연산기호 버튼만 눌러서 자동으로 값을 계산해 주는 계산기도 일종의 인공지능이라고 볼 수 있다.
# 일일이 셈을 계산하지 않고도 자동으로 계산을 도와주기 때문이다.
# 요즘 회자되는 AI는 좁은 의미로 딥러닝(Deep Learning)기반의 인공지능을 일컫는 경우가 대부분이다

# AI : 사람이 해야 할 일을 기계가 대신할 수 있는 모든 자동화에 해당

# Machine Learning : 명시적으로 규칙을 프로그래밍하지 않고 데이터로부터 의사결정을 위한 패턴을 기계가 스스로 학습

# Deep Learning : 인공신경망(Artificail Neural Network)기반의 모델을 활용해 
#                 비정형 데이터로부터 알아서 특징을 추출 및 판단까지 기계가 한 번에 수행하는
#                 기계학습(Machine Learning)의 한 방식이다.

# 인공신경망은 기계학습의 하위 기법 중 하나이지만, 빅데이터 붐이 일면서 유행한 
# 회귀 분석(Regression Analysis), 의사결정 나무(Decision Tree) 등의 전통적인 머신러닝 기법과는 조금 차이가 있다.

# 넓은 의미의 AI : 인간의 행위를 기계가 대신해 자동화 할 수 있는 모든 종류의 프로그램
# 좁은 의미의 AI : 딥러닝(인공신경망)을 기반으로, 데이터를 통해 의사결정에 필요한 패턴을 기계가 스스로 학습해
#                 인간의 행위를 자동화

# 전통적인 머신러닝 기법은 주로 정형 데이터를 다룬다. 관계형 데이터베이스(Relational Database)나 엑셀 표로 정리되는
# 테이블 데이터를 생각하면 된다. 의사결정에 필요한 데이터를 사람이 정리해 기계에 알려줌녀 기계는 이 정보를 토대로
# 판단이나 예측을 하는 경우이다.
# 반면 딥러닝은 주로 비정형 데이터를 다룬다. 비정형 데이터란 지정된 방식으로 정리됮 않은 정보를 말하는데
# 이미지, 비디오, 텍스트 문장이나 문서, 음성 데이터 등을 말한다.

# 인간이 가지고 있는 강아지와 고양이에 대한 지식을 기계에게 전수할 방법을 생각해보자.
# 강아지와 고양이를 구별할 수 있는 주요 특징(Feature)은
# 고양이는 대체로 귀가 작고 뾰족, 주둥이 길이가 강아지보다 짧은 특징, 박스 안에 들어가려고 하고
# 강아지는 박스를 물어뜯기를 좋아한다 등.
# '귀가 빳빳하고 각도가 60 이하면 고양이이다', '주둥이 길이가 5cm 이하면 고양이이다', 
# '박스를 줬을 때 들어가면 고양이이다' 같은 규칙들을 if-else구문으로 짜서, 기계에 눈앞의 생물에 대한 귀 각도,
# 주둥이 길이, 박스 선호도를 전달 인자로 주면 단순한 형태이자 넓은 의미에서의 인공지능을 만들 수 있다.
# 그러나 강아지처럼 귀가 접힌 스코티시폴드, 박스를 좋아하는 시베리안허스키, 모든 특징이 가려진 고양이 녀석 등
# 이러한 예외가 언제 어떤 케이스들에 발생할 지 알 수 없고, 그 때마다 매번 규칙을 수정하거나 추가해주는 유지보수에
# 공수가 많이 들어가는 문제가 발생한다.
# 그러면 사람은 어떻게 강아지와 고양이를 잘 분류할 수 있을까?

# '모라벨의 역설'이라는 개념이 있다. 인간에게 쉬운 일이 기계에게는 어렵고, 기계에게는 쉬운 일은 인간에게 어렵다는 말이다.
# 강아지와 고양이의 구별, 손을 뻗어 물체 집기, 걷고 뛰기, 지인들의 얼굴 구별하기, 자동가입 방지를 위한 꼬부랑 글자 식별 등
# 이러한 행위는 기계가 수행하기 어렵다. 반면 금융 시장에서 주가 시계열 예측, 어려운 학술지 번역, 복잡한 미적분 계산 등
# 이런 것들은 사람이 시간을 들여서 어렵게 수행해야 하지만 기계는 눈깜짝할 새 계산할 수 있다.

# 우리는 어릴 때 사전적 정의나 규칙이 아닌 다양한 여러 사례들을 보고 귀납적으로 강아지와 고양이에 대한 정보를 습득한다.
# 말로는 구체화하기 어렵지만, 대충 강아지와 고양이는 각각 이러저러한 특징들이 있다는 것을 깨닫는다.
# 그렇다면 기계에게도 수많은 사례를 통해 구별하게 되면 다양한 예외에도 잘 대처할 수 있지 않을까?
# 딥러닝은 바로 이러한 모티브를 기계학습에 녹인 것인다.
# 사람의 사고방식을 기계에 적용하기 위해, 이제부터 컴퓨터공학과 신경과학이 손을 잡는다.
# 사람에게는 '뉴런(Neuron)'이라는 신경 세포가 있어서, 세포의 한 쪽에서 받아들인 전기 자극 정보를
# 다른 쪽 끝에서 다음 세포로 전달하곤 한다. 이 모양을 본따서 만든 것이 '인공 뉴런(Artificial Neuron)'으로,
# 노드로부터 받아들인 데이터를 가중합 연산을 한 뒤 비선형함수를 적용하여 정보를 가공한 뒤, 다음 인공 뉴런으로
# 넘기는 역할을 한다. 이런 인공 뉴런을 다양한 방식으로 여러 층 쌓아 연결하게 되면
# 딥러닝의 기본 구조인 '인공신경망(Artificical Neural Network)'이 된다. 그리고 수많은 강아지와 고양이 사진을
# 인공신경망에게 보여주면 자동으로 강아지, 고양이 분류를 위한 최적의 연산 모델을 찾아낸다.
# 여기엔 귀가 어떻다든지, 주둥이 길이가 어떻다든지 하는 인간의 지식을 필요 없게 된다. 마치 우리가 어릴 때 학습한 것처럼.
# 이러한 딥러닝 모델은 '데이터를 통해' 자동으로 필요한 특징(말로는 표현하기 어렵지만)을 찾아내고 분류를 수행한다.

# 예전엔 어떻게 해서든지 인간이 알고 있는 지식(규칙)을 기계에 전수하려고 했다면,
# 이제는 인간이 사고하는 방식 그 자체를 기계에게 알려주고 데이터를 제공하는 입장이 되었다.
# 그 뿐만 아니라 최근 발전하는 '자동화 머신러닝 AutoML(Automated Machine Learning)'이라는 기술은 
# 인공신경망 구조 자체를 기계가 알아서 디자인하기도 한다.
# (모든 비즈니스에서 쉽게 AI를 사용할 수 있도록 돕는 클라우드 서비스이다. AI시스템을 구축할 수 있게 도와준다.)
# (한 마디로 머신러닝이 가능한 AI를 구축하는 AI인 것, 바야흐로 인공지능을 인공지능으로 만드는 시대가 도래하는 것이다.)
# 사람으로 따지면 뇌를 구성하는 최적의 방법 자체를 기계가 스스로 찾아나가는 것이다.

# 딥러닝 방식은 규칙 없이도 다양한 예외 케이스를 처리할 수 있지만, 쉽게 적용할 수 있는 것은 아니다
# 데이터로부터 특징을 스스로 찾아야 하니 양질의 많은 데이터를 확보해야만 높은 성능을 얻을 수 있고,
# 복잡해지는 인공신경망 구조의 연산을 수행해야 함에 따라 높은 스펙의 하드웨어를 필요로 한다.
# 최근 딥러닝이 화두가 되는 것은 이 두 가지의 제약조건을 해결할 방법이 점점 생겨나고 있기 때문이다.
# 그래서 많은 기업들이 데이터의 중요성을 깨닫고 양질의 데이터를 다량 확보하는 데 출력을 기울이고 있다.
# 그 중에서도 특히 비정형 데이터의 비중이 높아지고 있다.
# 규칙 기반이나 머신러닝 기반의 방식에 비정형 데이터를 이용하려면 태스크에 필요한 특징을 발라내는 작업이 필요한데,
# 딥러닝 방식은 이미지나 동영상, 텍스트 자료를 그대로 이용할 수 있다.
# 그런가하면 더 많은 데이터를 저장하고, 더 빠른 연산을 도와주는 하드웨어의 발전도 뒷받침을 해주는 요인이다.
# 1950년대에는 사람크기만한 서버에 5MB를 저장하는 하드디스크뿐이었지만, 요즘은 누구나 집에서 몇 TB씩을 손바닥만한
# 외장하드에 저장할 수 있다. 뿐만 아니라 GPU나 TPU를 활용하기 위해서도 비싼 서버 장비를 구매할 필요 없이
# 클라우드 서비스 플랫폼을 통해 필요한 만큼만 이용할 수 있는 등, 장비 접근성도 용이해졌다.

# 이러한 환경의 변화에 따라 성능이 좋은 딥러닝 알고리즘도 계속 발전하고 있으니, 모라벡의 역설이 바뀔 날도 오지 않을까?
# 기계에게 쉬운 일이 사람에게는 어려울 수 있다. 사람에게 쉬운 일은 기계에게도 곧, 쉬워질 것이다.

```

```python

# Python에서는 주석이 # 이다. (Java에서는 // 또는 /* */ 이거 였지)
# Python의 기본 자료구조
# list []        mutable   자료에서 수정/삽입/삭제가 가능하다. (인덱싱 가능)  *** 가장 많이 사용된다.
# tuple ()       immutable 자료에서 수정/삽입/삭제가 불가능하다. (인덱싱 가능)
# dictionary {}  mutable   Key와 Value로 이뤄진 자료형, 수정/삽입/삭제가 가능하다. (그러나 순서가 없다. 인덱싱 불가능) ** 데이터분석 시 사용된다.
# set {}         mutable   수정/삽입/삭제가 가능하다. 중복을 허용하지 않는다. (순서가 없다. 인덱싱 불가능)

# 실제 데이터 분석 시에는 '판다스 라이브러리'가 필수이다. (데이터 분석 필수 라이브러리)
# Pandas란??
# 파이썬에서 사용하는 데이터분석 라이브러리로, 행과 열로 이루어진 데이터 객체를 만들어 다룰 수 있게 되며
# 보다 안정적으로 대용량 데이터들을 처리하는데 편리한 도구이다.
# pandas를 먼저 설치해야 사용할 수가 있는데 설치한 이후 import pandas as pd (pandas 라이브러리를 import하고 그것을 pd라고 부르겠다는 의미)
# 보통 pandas 라이브러리를 부를 때 아래와 같이 부른다.
import pandas as pd  # 데이터 분석
import numpy as np   # 데이터 계산 (사용하기 편리하게 np로 부르겠다는 의미)


# Pandas 자료구조
# 기본적으로 Series와 DataFrame을 사용한다. 이 자료 구조들은 빅 데이터 분석에 있어서 높은 수준의 성능을 보여준다.
# Series (1차원 자료구조 / 행),           DataFrame (2차원 자료구조 / 행과 열)

# Series 정의하기
obj = pd.Series([4,7,-5,3])   
obj         # 0인덱스부터 시작해서 값이 행에 따라 출력된다.
obj.values  # 값(Values)만 확인할 수 있다.
obj.index   # 인덱스 확인 (0인덱스부터 시작, stop=요소갯수, step=인덱스1씩)
obj.dtypes  # 자료형 확인
```

```python
## 머신러닝
# 지도학습(Supervised learning) : 문제와 답을 함께 제시 (비슷한 문제가 왔을 때 해결할 수 있도록. 일본에서는 교사학습이라고도 한다.)
#                                이미지 분류, 양불판정, 감성 분석 등 우리가 생각하는 '인공지능에 시킬 수 있는 일' 대부분이
#                                이러한 지도 학습 방식으로 학습된다.
# 비지도학습(Unsupervised learining) : 문제만 제시 (컴퓨터가 스스로 답을 찾아야하기 때문에 문제의 양이 어마어마할 수 밖에 없다. 대표적으로 딥러닝)
#                                      정답 라벨 없이 이루어지는 알고리즘을 말한다. 이미지를 비슷한 것들끼리 묶는 군집 분석이
#                                      비지도 학습의 대표적인 예이다.
# 자기 지도 학습 : 라벨이 없는 데이터를 활용해 지도학습의 방식으로 학습하는 알고리즘을 말한다.
#                 사람이 데이터에 대한 라벨을 제공해주지 않아도 기계가 스스로 데이터와 라벨을 만들어 학습하면서 점점 똑똑해지는 것이다.


# 지도학습의 대표적인 유형 
# 회귀 : 데이터를 점으로 나타내고, 그 데이터들의 중간에 선을 긋는다. 점들이 선을 따라 움직이는 모양을 보인다. (주식, 온도, 집값 등)
# 분류 : 붓꽃(iris)분류, Versicolor, Setosa, Virginica 라는 서로 닮은 세 종류의 붓꽃 품종을 분류한다.

# 머신러닝 프로세스
# 문제파악 - 데이터 탐색 - 데이터 전처리(가장중요,전체분석에 70~80%차지,아직은 컴퓨터가 잘하지 못해 컴퓨터와 사람이 같이 하고 있다.)
#  - 모델학습 (머신러닝은 수학이기 때문에 컴퓨터가 잘 이해할 수 있도록, 남자/여자를 숫자로 변환해서, 모델을 계속해서 발전시키면 예측이 향상되겠지)
#  - 예측(결국 이 예측을 하기 위해, 집값을/주식을/비트코인을...)
```

```python
# 데이터 셋 분할
# 모델학습에 사용할 훈련(train) 데이터와 모델 성능을 평가하는데 사용할 테스트 데이터(test)를 분할한다.
# 보통 훈련용은 80%, 테스트는 20% (분석하는 사람에 따라 다르게 설정하는데 데이터양이 많으면 9:1 등.) 
# 현재 전체 데이터 150개에서 테스트용으로 30개, 나머지 훈련용 120개

from sklearn.model_selection import train_test_split    # 훈련용과 검증용으로 잘라줄거야
X_data = df.loc[:,'sepal_width':'petal_width']      # sepal_width 부터 petal_width 까지
y_data = df.loc[:,'Target']  # 전체 행에서, 'Target'만
# 관용적으로 x는 대문자로, y는 소문자로 사용한다.
# 컴퓨터에서 대문자는 바뀌지 않는 상수의 개념으로 쓰이고는 한다.
# X는 문제지를 훈련용과 검증용으로 나뉘고,
# y는 답안지를 훈련용과 검증용으로 나누다.
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.2, shuffle=True, random_state=20)
# random_state 는 항상 동일한 랜덤으로 출력되도록 한다. (매번 바뀌는 게 아니라)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# KNN 알고리즘    어떤식으로 학습시킬 것인지
# K-Nearest-Neighbors   K는 근접이웃
# k값에 따라 모델의 예측력이 달라지므로 적절한 k값을 설정해야 한다.

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=20)   # 생성자에 7을 던져주고(값을 바꿔보자)

knn.fit(X_train, y_train)  # 훈련용과 답안지를 준다.

y_knn_pred = knn.predict(X_test)   # 이제 문제만 던져주고, 

from sklearn.metrics import accuracy_score
knn_acc = accuracy_score(y_test, y_knn_pred)    # y_test(실제정답지), y_knn_pred(컴퓨터가 푼것)
knn_acc

# n = 7 , 0.9666666666666666666666667  이렇게 결과가 나왔다.
# n = 2 , 0.9                          정확도가 떨어진 것을 확인할 수 있다.
# n = 1 , 0.933333333333333333333      n값에 따라 비례하지 않는다는 것을 알 수 있다.
# n = 20 , 1.0                         100% 정확도를 보여준다.
# n = 50 , 0.9333333333333333333333    93%,정확도가 떨어졌다.
# n = 120 , 0.2666666666666            정확도가 많이 떨어졌다.
```

```python
# SVM Support Vector Machine 알고리즘
# 각 군집간의 거리(margin)를 최대한 멀리 유지하는 경계면을 찾는다.

from sklearn.svm import SVC
svc = SVC(kernel='rbf')   # 언제나 'rbf'(곡선)
svc.fit(X_train, y_train)  # 문제, 답

y_svc_pred = svc.predict(X_test)
svc_acc = accuracy_score(y_test, y_svc_pred)
svc_acc              # 1.0  100%가 나온다.
```

```python
# Logistic Regression 알고리즘
# 각 품종에 속하는 확률을 계산하여 1에 가까우면 그 클래스로 분류하고
# 0에 가까우면 아니라고 분류한다.

from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression()
lrc.fit(X_train, y_train)   # 문제와 답을 학습시킨다.

y_lrc_pred = lrc.predict(X_test)  # 문제만 준다.
lrc_acc = accuracy_score(y_test, y_lrc_pred)  # 실제답, 컴퓨터가 푼 답
lrc_acc                # 0.933333  93.33% 
```

```python
# 의사결정나무 decision tree 알고리즘
# 트리의 각 분기점에는 데이터셋의 피처(feature)를 하나씩 위치시켜서 해당 피처에 대한 임의의 조건식을 갖고
# 계속해서 2개의 줄기로 가지를 나누며 데이터를 구분한다.

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth = 1, random_state=20)   # random을 고정시킨다.
dtc.fit(X_train, y_train) # 문제와 답으로 학습을 시킨다.

y_dtc_pred = dtc.predict(X_test) # 테스트를 위해 문제만 준다.
dtc_acc = accuracy_score(y_test, y_dtc_pred)
dtc_acc                        # 0.9333 93.33% 
```

```python
## 앙상블 - 보팅, 배깅

# 보팅 - 하나의 모델 예측에 의존하는 것보다 여러 모델(알고리즘)의 예측을 종합하면 예측력이 일반적으로 더 좋다.

from sklearn.ensemble import VotingClassifier                            # 옵션은 'hard(한/두 표, 표로 결정)'와 'soft'가 있다
hvc = VotingClassifier(estimators = [('KNN',knn),('SVM',svc),('DT',dtc)],voting='hard')
hvc.fit(X_train, y_train)   # 문제,답을 넣는다.
y_hvc_pred = hvc.predict(X_test) 
hvc_acc = accuracy_score(y_test, y_hvc_pred)
hvc_acc    # 1.0      응? 모든 알고리즘을 합쳤는데 더 낮게 나오네..
```

```python
# 배깅 
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50, max_depth=5, random_state=20)
rfc.fit(X_train, y_train)
y_rfc_pred = rfc.predict(X_test)
rfc_acc = accuracy_score(y_test, y_hvc_pred)  # 채점
rfc_acc
```

```python
# 앙상블 부스팅
# XGBoost 모델은 캐글(kaggle), 데이콘(dacon) 등 경진대회에 가장 많이 사용되는 알고리즘이다.
# 모델 학습 속도가 빠르고 예측력이 상당히 좋은 편으로 알려져 있다.
# 여러 개의 모델을 순차적으로 학습한다. 오차를 줄일 수 있는 방향으로 모델을 계속 업데이트 한다.
# 여러 모델을 동시에 학습하지 않고 순차적으로 학습하는 점에서 다르다.

        배깅
        결론
    예측값을 종합
예측    예측    예측
모델1   모델2   모델3





                       부스팅
      
모델1예측 -> 모델2예측 -> 모델3예측(계속 업데이트) -> 최종 예측
```

```python
from xgboost import XGBClassifier
xgbc = XGBClassifier(n_estimators=50, max_depth=3, random_state=20)  # 학습시킬 변수를 생성하고 (기준이 될 파라미터 값을 넣어주고)
xgbc.fit(X_train, y_train)  # 문제와 답으로 학습시키고
y_xgbc_pred = xgbc.predict(X_test) # 테스트를 보고
xgbc_acc = accuracy_score(y_test, y_xgbc_pred)   # 채점
xgbc_acc         # 0.933    93% 정확도
```

```python
# 머신러닝을 돌리면서 이 문제들이 발생할 수 있다.
# 과대적합 overfitting  : 훈련 데이터와 관련된 예측은 정확하게 맞추지만 조금 다른 데이터는 제대로 인식하지 못하는 현상
#                         공을 학습시킬 때 축구공 위주로 학습시켜서 축구공은 잘 맞추는데, 다른 공 종류는 못찾는 그러한 현상
# 과소적합 underfitting : 훈련 데이터가 충분하지 못해서 예측을 제대로 하지 못하는 현상 

# 이 문제를 해결하기 위해 Hold out으로 데이터를 뻥튀기
```

```python
# Hold - out 교차검증

# X_train, y_train 훈련용 문제, 훈련용 정답 (80% 비율)
# X_test, y_test   검증용 문제, 검증용 정답 (20% 비율)

# 훈련용을 다시 자르자. 4개로 잘라보자.
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train,
                                            test_size = 0.3,     # 30%
                                            shuffle = True,
                                            random_state=20)

rfc = RandomForestClassifier(max_depth=3, random_state = 20)
rfc.fit(X_tr, y_tr)

y_tr_pred = rfc.predict(X_tr)
y_val_pred = rfc.predict(X_val)

tr_acc = accuracy_score(y_tr, y_tr_pred)     # 훈련용 
val_acc = accuracy_score(y_val, y_val_pred)  # 검증용

print(tr_acc, val_acc)    # 1.0 0.9166666666666666
# tr_acc 는 100%,  val_acc는 91.6%의 정확도가 나왔다. 이게바로 과대적합, 약 10% 차이가 난다.
# 검증용 정확도가 0.91로 낮은 수준이다. 훈련용 데이터에 과대적합하여 새로운 데이터에 대한 예측력이 떨어지고 있다.
```

```python
# k-fold (Hold-out이 확대된 것)
# Hold out은 학습 데이터를 다시 훈련용과 검증용으로 한 번만 분할하여 성능을 검증한다.
# k-fold는 hold out을 여러 번 반복하는 방법이다. 데이터를 서로 겹치지 않게 분할하여 모두 k번 검증하므로
# 모델의 일반화 성능을 평가하는데 뛰어나다.

from sklearn.model_selection import KFold
kfold = KFold(n_splits = 5, shuffle = True, random_state = 20)

num_fold = 1

for tr_idx, val_idx in kfold.split(X_train):
  print("%s Fold ------------------------------" % num_fold)
  print("훈련 : ", len(tr_idx), tr_idx[:23])     #[:10] 처음부터 10까지
  print("검증 : ", len(val_idx), val_idx[:23])
  num_fold += 1


```
